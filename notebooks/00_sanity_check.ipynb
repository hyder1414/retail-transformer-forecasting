```python
from pathlib import Path

RAW = Path("../data/raw")

def head_text(path, n=3, clip=200):
    print(f"\n### {path.name}")
    with open(path, "r") as f:
        for i, line in zip(range(n), f):
            print(line[:clip] + ("â€¦" if len(line) > clip else ""))

for name in ["sales_train_validation.csv", "calendar.csv", "sell_prices.csv"]:
    head_text(RAW / name)

```
